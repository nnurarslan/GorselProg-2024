```python
# Aylık elektrik faturaları
faturalar = [120, 180, 135, 160, 210, 145]

# Fatura durumlarını belirten boş liste
fatura_durumlari = []

# Fatura kontrolü
for fatura in faturalar:
    if fatura > 150:
        fatura_durumlari.append("Yüksek Fatura")
    else:
        fatura_durumlari.append("Normal Fatura")

# Sonuçları yazdırma
for i in range(len(faturalar)):
    print(f"Ay {i+1} - Fatura: {faturalar[i]} TL, Durum: {fatura_durumlari[i]}")
```

```python
# Kullanıcıdan kaç not gireceğini öğreniyoruz
not_sayisi = int(input("Kaç tane not gireceksiniz? "))

# Notları tutmak için boş bir liste oluşturuyoruz
notlar = []

# Kullanıcıdan notları girmesini istiyoruz
for i in range(not_sayisi):
    notu = int(input(f"{i+1}. notu girin: "))
    notlar.append(notu)

# Geçti veya Kaldı durumlarını belirleyen boş bir liste
durumlar = []

# Her not için kontrol yapıyoruz
for notu in notlar:
    if notu >= 50:
        durumlar.append("Geçti")
    else:
        durumlar.append("Kaldı")

# Sonuçları yazdırıyoruz
for i in range(len(notlar)):
    print(f"Öğrenci {i+1} - Not: {notlar[i]}, Durum: {durumlar[i]}")
```

```python
# Ürün adlarını ve stok durumlarını tutan dictionary
urunler = {}

# Kaç ürün olduğunu kullanıcıdan alıyoruz
urun_sayisi = int(input("Kaç ürün gireceksiniz? "))

# Kullanıcıdan ürün adı ve stok bilgisi alıyoruz
for i in range(urun_sayisi):
    urun_adi = input(f"{i+1}. ürünün adını girin: ")
    stok = int(input(f"{urun_adi} ürününün stok miktarını girin: "))
    urunler[urun_adi] = stok

# Stok durumu kontrolü ve yazdırma
for urun, stok in urunler.items():
    if stok > 0:
        print(f"{urun} - Stok: {stok} - Durum: Mevcut")
    else:
        print(f"{urun} - Stok: {stok} - Durum: Tükendi")
```

```python
# Öğrenci notlarını ve kulüp katılımlarını tutan dictionary'ler
ogrenci_notlari = {}  # Anahtar: Öğrenci adı, Değer: Notlar listesi
ogrenci_kulupleri = {}  # Anahtar: Öğrenci adı, Değer: Katıldığı kulüpler listesi

# Kaç öğrenci olduğunu kullanıcıdan alıyoruz
ogrenci_sayisi = int(input("Kaç öğrenci gireceksiniz? "))

# Öğrenci bilgilerini kullanıcıdan alıyoruz
for _ in range(ogrenci_sayisi):
    ad = input("Öğrencinin adını girin: ")
    
    # Eğer öğrenci daha önce eklenmediyse, notlar ve kulüpler için boş liste oluştur
    if ad not in ogrenci_notlari:
        ogrenci_notlari[ad] = []
    if ad not in ogrenci_kulupleri:
        ogrenci_kulupleri[ad] = []
    
    # Her öğrencinin 3 farklı sınav notunu alıyoruz
    for i in range(1, 4):
        notu = int(input(f"{ad} öğrencisinin {i}. notunu girin: "))
        ogrenci_notlari[ad].append(notu)
    
    # Öğrencinin katıldığı kulüpleri alıyoruz
    kulupler = input(f"{ad} öğrencisinin katıldığı kulüpleri (virgülle ayırarak) girin (Boş bırakabilirsiniz): ").split(',')
    
    # Eğer boşsa katılım yoktur
    if kulupler != ['']:  
        ogrenci_kulupleri[ad].extend([kulup.strip() for kulup in kulupler])

# 1. Aşama: Öğrencilerin geçip kalma durumunu hesapla
print("\n--- Öğrenci Geçme/Kalma Durumu ---")
for ad, notlar in ogrenci_notlari.items():
    ortalama = sum(notlar) / len(notlar)
    if ortalama >= 50:
        print(f"{ad} - Ortalama: {ortalama:.2f} - Durum: Geçti")
    else:
        print(f"{ad} - Ortalama: {ortalama:.2f} - Durum: Kaldı")

# 2. Aşama: Kulüp katılım analizini yap
print("\n--- Kulüp Katılım Analizi ---")
# Kulüp katılımlarını saymak için bir dictionary oluşturuyoruz
kulup_sayaci = {}

# Her öğrencinin katıldığı kulüpleri sayıyoruz
for ad, kulupler in ogrenci_kulupleri.items():
    # Eğer öğrenci hiçbir kulübe katılmadıysa sıfır yazdır
    if not kulupler:
        print(f"{ad} hiçbir kulübe katılmadı: 0 katılım")
    else:
        for kulup in kulupler:
            if kulup in kulup_sayaci:
                kulup_sayaci[kulup] += 1
            else:
                kulup_sayaci[kulup] = 1

# En çok katılım gösterilen kulübü bulma
if kulup_sayaci:
    en_populer_kulup = max(kulup_sayaci, key=kulup_sayaci.get)
    print("\nEn çok katılım gösterilen kulüp:", en_populer_kulup, "-", kulup_sayaci[en_populer_kulup], "katılım")

print("\n--- Tüm Kulüpler ve Katılım Sayıları ---")
for kulup, sayi in kulup_sayaci.items():
    print(f"{kulup}: {sayi} katılım")

# 3. Aşama: Öğrencilerin katıldığı kulüpleri listele
print("\n--- Öğrenci Bazlı Kulüp Katılımı ---")
for ad, kulupler in ogrenci_kulupleri.items():
    if not kulupler:
        print(f"{ad} hiçbir kulübe katılmadı.")
    else:
        print(f"{ad} şu kulüplere katıldı: {', '.join(kulupler)}")
```
